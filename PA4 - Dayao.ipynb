```python
#load the .xlsx file into a data frame

board = pd.read_excel('board2.xlsx')
board
```

```python
#acquire average grade

board['Average'] = board[['Electronics','GEAS','Math','Communication']].mean(axis=1)
```

**PROBLEM 1**<br>
Create the following data frames based on the format provided.

```python
#example: Vis = [“Name”, “Gender”, “Track”, “Math<70”]; hometown is constant as Visayas 

Vis = board.loc[(board['Math']<70) & (board['Hometown']=='Visayas'),['Name','Gender','Track','Math']]
Vis
```

**PROBLEM A**<br>
Data frame 'Instru' showing 'Name', 'GEAS', and 'Electronics' grades greater than 70; Instrumentation and Luzon are constant as track and hometown, respectively.

```python
#a. Filename: Instru = [“Name”, “GEAS”, “Electronics >70”]; where track is constant as Instrumentation and hometown Luzon

Instru = board.loc [(board['Electronics']>70)
    & (board['Track']=='Instrumentation')
    & (board['Hometown']=='Luzon'),
    ['Name','GEAS','Electronics']]

Instru
```

**PROBLEM B**<br>
Data frame 'Mindy' showing 'Name', 'Track', and 'Electronics', and 'Average' grades greater than or equal to 55; Mindanao and Female are constant as hometown and gender, respectively.

```python
#b. Filename: Mindy = [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is constant as Mindanao and gender Female

Mindy = board.loc [(board['Average']>=55)
    & (board['Hometown']=='Mindanao')
    & (board['Gender']=='Female'),
    ['Name','Track','Electronics','Average']]

Mindy
```

**PROBLEM 2**<br>
Create a visualization that shows how the different features contributes to average grade. Does 
chosen track in college, gender, or hometown contributes to a higher average score?

```python
#create a visualization according to Average by Track, Average by Gender, and Average by Hometown
fig, axes = plt.subplots(3, 1, figsize=(7.5, 12))

#bar graph for average grade by track
board.groupby('Track')['Average'].mean().plot(kind='bar', ax=axes[0], color='#EAB0C3', title='Average Grade by Track', ylabel='Average Grade')

#bar graph for average grade by gender
board.groupby('Gender')['Average'].mean().plot(kind='bar', ax=axes[1], color='#AE93D9', title='Average Grade by Gender', ylabel='Average Grade')

#bar graph for average grade by hometown
board.groupby('Hometown')['Average'].mean().plot(kind='bar', ax=axes[2], color='#F0D485', title='Average Grade by Hometown', ylabel='Average Grade')

#print combined plot
plt.tight_layout()
plt.show()
```
